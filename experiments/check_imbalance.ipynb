```python
folder_A = "C:\Projects Datasets\Segmentation_1\Segmentation\\images"
folder_B = "C:\Projects Datasets\Segmentation_1\Segmentation\\masks"

# Get filenames only
files_A = set(os.listdir(folder_A))
files_B = set(os.listdir(folder_B))

# Files in A but NOT in B
files_to_remove = files_A - files_B

print(f"Files to be removed: {len(files_to_remove)}")

for filename in files_to_remove:
    file_path = os.path.join(folder_A, filename)

    # Make sure it's a file (not a folder)
    if os.path.isfile(file_path):
        os.remove(file_path)
        print(f"Removed: {filename}")
```
